Why do computers need NLP to understand human language? Give one real-world
example to support your answer.

Computers need NLP because human language is complex and ambiguous, unlike the structured data they typically process. NLP allows them to interpret meaning.
Example: A chatbot needs NLP to understand customer queries like "I want to track my order" and respond appropriately.



List any four real-life applications of NLP and briefly describe what each one does.




1. Spam Detection: Identifies unwanted emails by analyzing text for suspicious patterns.
2. Sentiment Analysis: Determines the emotional tone (positive, negative, neutral) in text data.
3. Machine Translation: Translates text from one human language to another (e.g., Google Translate).
4. Chatbots/Virtual Assistants: Understands and responds to user commands or questions in natural language.

What is a rule-based NLP system? Explain using the spam detection example
(checking if "Congratulations" is present in a message).

A rule-based NLP system uses predefined linguistic rules or patterns to process text. It doesn't learn from data.
For spam detection, a rule could be: if the word "Congratulations" is in the message, label it as spam.

List and briefly explain the seven steps of the NLP pipeline, in the correct order.

1. Tokenization: Breaking text into smaller units (words/sentences).
2. Lowercasing: Converting all text to a uniform lowercase format.
3. Stopword Removal: Eliminating common, less meaningful words (e.g., "the", "is").
4. Stemming/Lemmatization: Reducing words to their root form.
5. Part-of-Speech Tagging: Labeling words with their grammatical role.
6. Named Entity Recognition: Identifying proper nouns (people, places, organizations).
7. Dependency Parsing: Analyzing grammatical relationships between words.

What is tokenization? Why is it a necessary step before a machine learning model
can process text?

Tokenization is the process of splitting text into smaller units called tokens (words, phrases, symbols).
It's necessary because ML models require numerical input, and tokenization breaks text into discrete items that can be converted into numerical representations.

What are stopwords? Give three examples of stopwords in English and explain why
they are usually removed

Stopwords are common words that appear frequently but often carry little semantic meaning. Examples: "the", "is", "and".
They are removed to reduce noise and dimensionality in text data, improving efficiency and focus on more important words.

Why is it important to convert all text data to lowercase before processing it in NLP?
Illustrate with an example.

Lowercasing standardizes text, treating words like "Apple", "apple", and "APPLE" as the same token. This prevents the model from treating them as distinct words.
Example: Without lowercasing, "Good" and "good" would be seen as different, leading to inconsistent analysis.

Explain the difference between using = and == while filtering rows in a DataFrame,
and why mixing them up causes errors.

`=` is for assignment, used to give a value to a variable or column. `==` is for comparison, checking if two values are equal.
Mixing them up causes errors because `df[df['col'] = 'value']` attempts assignment within a boolean indexing context, which is invalid for comparison.
