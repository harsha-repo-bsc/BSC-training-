```python
# stop words

from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.tokenize import word_tokenize # Ensure word_tokenize is imported if not already

stop_words = set(stopwords.words('english'))

stop_removed_sentences = []
for text in df['cleaned']:
  filtered_words = [word for word in word_tokenize(text) if word.lower() not in stop_words]
  stop_removed_sentences.append(' '.join(filtered_words))

df['stop_removed'] = stop_removed_sentences

df.head()
```

```python
# stemming
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize # Ensure word_tokenize is imported if not already

stemmer = PorterStemmer()

stemmed_sentences = []
for text in df['stop_removed']:
  stemmed_words = [stemmer.stem(word) for word in word_tokenize(text)]
  stemmed_sentences.append(' '.join(stemmed_words))

df['stemmed_text'] = stemmed_sentences

df.head()
```

```python
# lemmatize - dictioary word

import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize # Ensure word_tokenize is imported if not already

lemmatizer = WordNetLemmatizer()

lemmatized_sentences = []
for text in df['stop_removed']:
  lemmatized_words = [lemmatizer.lemmatize(word) for word in word_tokenize(text)]
  lemmatized_sentences.append(' '.join(lemmatized_words))

df['lemmatized_text'] = lemmatized_sentences

df.head()
```

```python
import re , string , nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer # Import necessary modules

# Download necessary NLTK data (quiet=True to suppress repeated messages if already downloaded)
nltk.download('stopwords', quiet=True)
nltk.download('wordnet', quiet=True) # Ensure wordnet is downloaded for lemmatization

# Initialize NLP tools
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

# Start with the 'Text' column for processing
# We'll create a new column to store the fully processed text, to keep original 'Text' intact
df['fully_processed_text'] = df['Text'].copy()

# Lowercasing
df['fully_processed_text'] = df['fully_processed_text'].str.lower()

# URL removal
df['fully_processed_text'] = df['fully_processed_text'].str.replace(r'http?://\S+|www\.\S+','', regex=True)

# Punctuation removal
# Define a translation table once
translator = str.maketrans('', '', string.punctuation)
# Apply translation to each string in the Series
df['fully_processed_text'] = df['fully_processed_text'].apply(lambda x: x.translate(translator) if isinstance(x, str) else x)

# Tokenization, Stop word removal, Stemming, and Lemmatization in a single apply statement
# This processes each text entry through the remaining steps sequentially.
df['fully_processed_text'] = df['fully_processed_text'].apply(lambda text_input:
    ' '.join([lemmatizer.lemmatize(stemmer.stem(word))
              for word in word_tokenize(text_input)
              if word.lower() not in stop_words])
    if isinstance(text_input, str) else ""
)

# Display the DataFrame head with the new column
df.head()
```

```python
from nltk.chunk import RegexpParser
sentence="the intelligent student won the competition"
tokens=word_tokenize(sentence)
tags=pos_tag(tokens)
grammer="NP: {<DT>?<JJ>*<NN>}"
cp=RegexpParser(grammer)
result=cp.parse(tags)
print(result)
```

```python
# Dependency Parsing

import spacy
from spacy import displacy

nlp = spacy.load("en_core_web_sm")
doc = nlp("John eats pizza.")

for token in doc:
    print(token.text, token.dep_, token.head.text)

displacy.render(doc, style="dep", jupyter=True)
```

```python
from spacy import displacy

nlp = spacy.load('en_core_web_sm')

doc = nlp('i love natural langugae process')

for entity in doc.ents:
    print(entity.text, entity.label_)

displacy.render(doc, style='ent', jupyter=True)
```
