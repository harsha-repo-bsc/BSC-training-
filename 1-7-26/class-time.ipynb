```python
dic = {}

class Books:
    Books = {
    'abc' : 123,
    'xyz' : 456,
    'pqr' : 789
                }
    
    def display_books(self):
        return self.Books


class Student(Books) :
    def __init__(self, name, id):
        self.name = name
        self.id = id

    

    def get_student_details(self):
        return f"Name: {self.name}, ID: {self.id}"
    
    def ask_book(self,book_name):
        if book_name in self.Books:
            dic[self.id] = book_name
            return f"{self.name} has borrowed the book: {book_name}"
        else:
            return f'sorry that book not available'
    def my_books(self):
        if self.id in dic:
            return f'{self.name} has borrowed the book {dic[self.id]}'
        
    


    
b = Books()
print('AVAILABLE BOOKS IN LIBRARY')
print(b.display_books())

s1 = Student('john', 1)
print(' STUDENT DETAILS')
print(s1.get_student_details())


print(s1.my_books())

print(s1.ask_book('abc'))
print(s1.my_books())
```
