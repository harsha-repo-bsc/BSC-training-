#### 1. Define a Large Language Model in your own words and explain what it means for a model to "learn the patterns of language."

A Large Language Model (LLM) is an AI model, typically based on deep learning architectures like Transformers, trained on vast amounts of text data. For a model to "learn the patterns of language" means it acquires statistical relationships, grammar, semantics, and context from this data, allowing it to predict subsequent words or generate coherent text.

#### 2. Explain why LLMs are called "large." Discuss both the training data dimension and the parameter dimension.

LLMs are called "large" due to their immense scale in two main aspects: the **training data dimension** (they are trained on petabytes of text and code from the internet) and the **parameter dimension** (they possess billions or even trillions of trainable parameters. This vastness enables them to capture complex language nuances and generalize across diverse tasks.

#### 3. Describe the evolution of language models from rule-based systems to modern LLMs, explaining why each stage improved on the previous one.

Language models evolved from **rule-based systems** (inflexible), to **statistical models** (e.g., n-grams, which learned probabilities but lacked long-range context), then to **neural networks** (RNNs/LSTMs, which captured sequential dependencies better), and finally to **modern LLMs** (Transformers, which excel at parallel processing and long-range context through attention mechanisms), each stage offering greater flexibility, context understanding, and performance.

#### 4. Explain the key limitations of RNNs and LSTMs that motivated the invention of the Transformer architecture.

RNNs and LSTMs suffered from **sequential processing bottlenecks**, making parallelization difficult and slowing training on long sequences. More critically, they struggled with **long-range dependencies**, as information could degrade over many steps, making it hard to connect distant words. The Transformer's self-attention mechanism addressed these issues by allowing direct, parallel connections between all words.

#### 5. What is self-attention, and how does it allow a Transformer to overcome the long-range dependency problem faced by older models?

Self-attention is a mechanism allowing a model to weigh the importance of all other words in a sequence when processing a single word, effectively creating direct connections between any two words regardless of their distance. This global view enables Transformers to directly capture long-range dependencies, overcoming the limitations of sequential models that struggled to maintain context over long distances.

#### 6. List and explain three real-world limitations of LLMs (e.g. hallucination, knowledge cutoff, context window), and suggest one way to mitigate each.

1.  **Hallucination**: LLMs can generate factually incorrect but plausible-sounding information. Mitigation: **Retrieval Augmented Generation (RAG)** by grounding responses in verified external data.
2.  **Knowledge Cutoff**: LLMs have limited knowledge up to their last training data. Mitigation: **Fine-tuning** with up-to-date information or using **real-time search integrations**.
3.  **Context Window**: Limited input length means LLMs can forget earlier information in very long texts. Mitigation: Employing **sliding window attention** or **hierarchical attention mechanisms**.

#### 7. Differentiate between open-weight and closed/hosted LLMs, giving at least two examples of each and comparing their trade-offs.

**Open-weight LLMs** (e.g., Llama 2, Mistral) have their model weights publicly accessible, allowing for full customization and local deployment, offering transparency and control but requiring more technical expertise. **Closed/hosted LLMs** (e.g., GPT-4, Claude) are proprietary models provided via API, offering ease of use and often higher performance out-of-the-box, but with less transparency and vendor lock-in.

#### 8. Explain why an LLM cannot process raw text directly and describe the role tokenization plays in bridging this gap.

LLMs cannot process raw text directly because they are mathematical models that operate on numerical data. **Tokenization** bridges this gap by converting raw text into discrete numerical units (tokens), which are then mapped to numerical embeddings. This process allows the LLM to understand and manipulate text as vectors in a high-dimensional space.

#### 9. Compare character-level, word-level, and subword-level tokenization, listing one advantage and one disadvantage of each.

-   **Character-level**: Advantage: Handles OOV words well. Disadvantage: Very long sequences, poor semantic understanding.
-   **Word-level**: Advantage: Directly maps to semantic units. Disadvantage: Large vocabulary, poor OOV handling.
-   **Subword-level** (e.g., BPE): Advantage: Balances vocabulary size and OOV handling. Disadvantage: Tokens might not align with linguistic units.

#### 10. Describe how Byte Pair Encoding (BPE) builds its vocabulary, using a short example of the merge process.

BPE builds its vocabulary iteratively by identifying the most frequent pairs of bytes (or characters/subwords) in the training data and merging them into new, single tokens. This process repeats until a desired vocabulary size or number of merges is reached.

Example: Given text "aaabdaaabac", BPE might first merge "aa" to `X`: "XabdXbac". Then if "ab" is frequent, it merges to `Y`: "XydXyac".

#### 11. Explain the difference between WordPiece and SentencePiece tokenization, including a scenario where each would be preferred.

**WordPiece** tokenization is typically used after an initial word segmentation (space-tokenization) and is language-dependent. It learns subword units from pre-segmented words. **SentencePiece** treats the entire input as a raw stream of characters, including whitespace, and tokenizes directly from it. This makes it language-agnostic and reversible. SentencePiece is preferred for multilingual models or when dealing with languages that don't use spaces for word separation.

#### 12. What are special tokens (e.g. `[CLS]`, `[SEP]`, `[PAD]`, `[UNK]`) used for, and why does a model need them?

Special tokens serve specific purposes in an LLM's input to provide structural information or handle edge cases. `[CLS]` marks the start of a sequence for classification, `[SEP]` separates segments, `[PAD]` fills shorter sequences to a fixed length, and `[UNK]` represents out-of-vocabulary words. Models need them to understand sequence boundaries, distinguish parts of input, maintain consistent input sizes, and gracefully handle unknown words, all critical for proper processing and task execution.

#### 13. Explain the purpose of an attention mask in the context of padded sequences.

An attention mask is used to prevent the self-attention mechanism from attending to padding tokens in a sequence. When sequences are padded to a uniform length, the mask typically contains zeros for padding tokens and ones for actual content tokens. This ensures that the model's attention scores are not skewed by irrelevant padding, allowing it to focus only on meaningful parts of the input and preventing information leakage from non-existent tokens.

#### 14. Explain why token IDs alone are not sufficient input for a Transformer, and how embeddings solve this problem.

Token IDs alone are discrete, categorical integers and carry no inherent semantic meaning or relationship between words. A Transformer needs rich, continuous representations of words to understand their meaning and context. Embeddings solve this by mapping each token ID to a dense vector in a high-dimensional space, where semantically similar words are located closer to each other, providing the model with meaningful numerical inputs.

#### 15. Compare one-hot encoding with dense embeddings in terms of memory efficiency and semantic meaning.

**One-hot encoding** represents each word as a sparse vector with a 1 at its index and 0 elsewhere, making it very memory inefficient for large vocabularies and lacking any semantic relationship between words (all vectors are orthogonal). **Dense embeddings** represent words as compact, continuous vectors, which are far more memory efficient and capture semantic meaning, allowing for relationships like 'king' - 'man' + 'woman' = 'queen'.

#### 16. Describe how the embedding layer works as a lookup table, and explain whether embeddings are meaningful before training begins.

The embedding layer works like a lookup table: given a token ID, it retrieves the corresponding dense vector from a matrix. Each row in this matrix represents the embedding for a specific token. Before training begins, these embeddings are typically initialized randomly, meaning they are not inherently meaningful. Their semantic richness and relationships are learned and refined during the model's training process as it processes vast amounts of text data.

#### 17. What is the difference between input embeddings and output embeddings in a Transformer, and what is "weight tying"?

**Input embeddings** convert input tokens into numerical representations for the model to process. **Output embeddings** convert the model's final hidden states back into a probability distribution over the vocabulary to predict the next token. "**Weight tying**" is a technique where the input embedding matrix and the output embedding matrix (or their transpose) share the same weights. This reduces the number of parameters and can improve generalization, especially in models with large vocabularies.

#### 18. Explain what positional embeddings are for, using the "Dog bites man" vs. "Man bites dog" example to justify your answer.

Positional embeddings provide Transformers with information about the order or position of tokens in a sequence. Without them, the self-attention mechanism treats all tokens equally regardless of their placement, losing critical sequential information. For "Dog bites man" vs. "Man bites dog," positional embeddings are crucial: they differentiate the subject and object, allowing the model to correctly interpret the meaning based on word order, which a bag-of-words approach cannot achieve.

#### 19. Explain contextual embeddings using the word "bank" as an example, and contrast this with traditional fixed embeddings.

**Contextual embeddings** generate a unique vector for a word based on its surrounding context in a sentence. For example, the word "bank" would have different embeddings in "river bank" versus "financial bank." In contrast, **traditional fixed embeddings** (like Word2Vec or GloVe) assign a single, static vector to each word regardless of its usage, failing to capture polysemy (multiple meanings of a word).

#### 20. Explain cosine similarity as a measure of semantic closeness between two embedding vectors, including what values of -1, 0, and 1 represent.

Cosine similarity measures the cosine of the angle between two embedding vectors, indicating their directional similarity and thus their semantic closeness. Values range from -1 to 1:
-   **1**: Indicates identical direction, meaning the vectors are semantically very similar.
-   **0**: Indicates orthogonality, meaning the vectors are unrelated in meaning.
-   **-1**: Indicates opposite direction, meaning the vectors are semantically opposite or highly dissimilar.

#### 21. Explain the roles of the Query, Key, and Value vectors in self-attention using the library analogy discussed in class.

In the library analogy, the **Query (Q)** is like your search query for books, the **Key (K)** is like the book's index/tags that match your query, and the **Value (V)** is the book itself (its content). Self-attention uses Q to query K, finding relevant V to create a new, context-aware representation.

#### 22. Walk through the steps of scaled dot-product attention, starting from Q, K, V and ending at the context vector.

1.  Calculate attention scores by taking the dot product of **Query (Q)** and **Key (K)**.
2.  Scale these scores by dividing by the square root of the key's dimension ($\sqrt{d_k}$).
3.  Apply a Softmax function to get attention weights.
4.  Multiply these weights by the **Value (V)** vectors and sum them up to produce the context vector, which is the output.

#### 23. Why is the dot product between Q and K scaled by $\sqrt{d_k}$ before applying softmax? What problem does this scaling solve?

The dot product values between Q and K can become very large, especially with high dimensions ($d_k$), pushing the Softmax function into regions with extremely small gradients. Dividing by $\sqrt{d_k}$ scales these values down, preventing the Softmax from saturating. This helps stabilize training by ensuring gradients remain in a reasonable range, allowing the model to learn more effectively.

#### 25. Describe how the embedding dimension is split across multiple heads, and how the outputs of each head are recombined.

The input embedding dimension is split into smaller dimensions, and each attention head operates independently on one of these smaller subspaces. After each head computes its own context vector, their outputs are concatenated back together. This combined output then passes through a final linear layer to project it back to the original embedding dimension, allowing the model to integrate diverse relational insights.

#### 26. Explain the purpose of residual connections and Layer Normalization within a Transformer block.

**Residual connections** (skip connections) help mitigate the vanishing gradient problem in deep networks by allowing gradients to flow directly through the network, aiding training convergence. **Layer Normalization** normalizes the inputs across the features for each sample, stabilizing the training process and speeding up convergence by keeping the activations within a stable range, regardless of the batch size.

#### 27. Describe the role of the Feed Forward Network (FFN) and Dropout within a Transformer block.

The **Feed Forward Network (FFN)** applies a position-wise, two-layer linear transformation to each token's representation independently, allowing the model to process information at a higher level of abstraction after attention. **Dropout** is a regularization technique that randomly sets a fraction of neurons to zero during training, preventing overfitting by forcing the network to learn more robust features.

#### 28. Compare encoder and decoder architectures, explaining why GPT uses only a decoder.

An **encoder** (e.g., in BERT) processes the entire input sequence to understand its full context, generating a rich representation. A **decoder** (e.g., in GPT) generates output sequences token by token, typically only attending to past tokens. GPT uses only a decoder because its primary task is text generation, which requires predicting the next token based on the preceding ones, making a causal (masked) self-attention mechanism essential.

#### 29. Explain masked (causal) self-attention and why it is necessary for next-token prediction.

Masked (causal) self-attention prevents a token from attending to future tokens in the sequence during training. It ensures that when the model is predicting the next word, it only uses information from words that have already occurred. This causal masking is crucial for next-token prediction tasks, as it mimics the sequential nature of language generation, preventing the model from 'cheating' by seeing the answer.

#### 30. Describe the autoregressive text generation loop in GPT, from an initial prompt to a completed response.

Starting with an initial prompt, GPT predicts the next most likely token. This newly predicted token is then appended to the prompt, and the updated sequence becomes the new input. This loop continues, with the model recursively predicting one token at a time based on all previously generated tokens and the original prompt, until a stop condition (e.g., max length, `EOS` token) is met, forming a complete response.

#### 31. Explain what a KV cache is and why it improves inference efficiency during text generation.

A **KV cache** (Key-Value cache) stores the computed Key and Value vectors from previous tokens in a sequence. During autoregressive generation, instead of recomputing these K and V pairs for every token from scratch, the model can retrieve them from the cache. This significantly improves inference efficiency by reducing redundant computations, making text generation much faster, especially for long sequences.

24. Write code that pads a batch of three sentences of different lengths to the same length and
prints the corresponding attention mask for each sentence

```python
# Define a few example texts of different lengths
batch_texts = [
    'This is a short sentence.',
    'This is a slightly longer sentence for demonstration purposes.',
    'Another text.'
]

# Tokenize each text in the batch
batch_tokenized = [tokenizer.encode(text) for text in batch_texts]

print("Original token IDs for each sequence:")
for i, ids in enumerate(batch_tokenized):
    print(f"Sequence {i+1} (length {len(ids)}): {ids}")

# Set the pad_token for the tokenizer. For GPT2, it's common to use the EOS token.
# This resolves the ValueError encountered previously.
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

# Identify the pad token ID. For GPT2, it's typically the EOS token ID.
pad_token_id = tokenizer.pad_token_id
print(f"\nPad token ID: {pad_token_id}")

# Pad the batch of tokenized sequences
# padding='longest' pads to the length of the longest sequence in the batch
# return_attention_mask=True generates an attention mask
padded_batch = tokenizer.pad(
    {'input_ids': batch_tokenized},
    padding='longest',
    return_attention_mask=True,
    return_tensors='pt' # Return as PyTorch tensors for convenience
)

print("\nPadded input_ids (token IDs):")
print(padded_batch['input_ids'])

print("\nAttention mask (1 for real tokens, 0 for padding tokens):")
print(padded_batch['attention_mask'])

print("\nDecoded padded input IDs to show the original and padding:")
for i, input_ids in enumerate(padded_batch['input_ids']):
    print(f"Sequence {i+1}: {tokenizer.decode(input_ids, skip_special_tokens=False)}")
```
