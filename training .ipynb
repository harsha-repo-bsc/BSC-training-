```python


# considering the list with deplicates of dictionary items , and  *** deduplicating them 

l = [ {'a':1} , {'b':2} , {'c':3} , {'a':1}]  # noqa: E741
l1 = []
uniques = set()
for i in l:
    t = (tuple(i.items()))
    if t not in uniques:
        uniques.add(t)
        l1.append(i)

print(uniques)
print(l1)
```

```python
l_comp = [ data.items() for data in l if 'a' in data.keys()  ]
print(l_comp)

```

```python
for data in l :
  
    if 'a' in data.keys() :
        print('matched key is :')
        print(data.items())
```

```python
# missing key function 

l = [ {'a':1} , {'b':2} , {'':3} , {'a':1}]  # noqa: E741


def missing_key( s ):
    missingkey = [data.items() for data in s if '' in  data.keys()]

    return missingkey

print(missing_key(l))
```

* list comprehension

```python
#    LIST COMPREHENSIONS

squares = [ {x:x**2} for x in range(5)]
squares
```

```python
even_squares = [ {x:x**2 for x in range(5) if x%2==0}]
even_squares
```

```python
l1 = [1,2,3,4,5]
l2 = ['a','b','c','d','e']

make_dic = [ {i:j} for i , j in zip(l2,l1) ]   # zip function to map 2 lists based on their index 
make_dic
```

* dictionary comprehensions

```python
dic = {'arun': 1, 'barath': 2, 'chandu': 3, 'dany': 4,'eshwar': 5}
swapped_dic = { value : key for key , value in dic.items()}

print(swapped_dic)
```

```python
# create uppercase keys or values 

upper_case = { value.upper() : key for key , value in swapped_dic.items()}
upper_case
```

```python
# conditional dictionaries

d = { x: { 'even' if x%2==0 else 'odd'} for x in range(5) }
d
```

*  Generators

```python
# INSTEAD OF RETURN IT YEILDS THE OUTOUT ONE AT A TIME WITHOUT USING MORE MEMORY

def gen():
    for i in range(16):
        yield i
a = gen()

print(next(a))
print(next(a))


```

```python
def gen():
    yield 10
    yield 20
    yield 30

a = gen()

print(next(a))
print(next(a))
print(next(a))






```
