```python
# considering the list with deplicates of dictionary items , and deduplicating them 

l = [ {'a':1} , {'b':2} , {'c':3} , {'a':1}]  # noqa: E741
l1 = []
uniques = set()
for i in l:
    t = (tuple(i.items()))
    if t not in uniques:
        uniques.add(t)
        l1.append(i)

print(uniques)
print(l1)
```
