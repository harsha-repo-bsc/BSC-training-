Hypothesis testing is a method used to check whether a claim or idea is supported by data.

Simple meaning:

A hypothesis is an assumption or guess.
Hypothesis testing is the process of testing that guess using evidence.

Chi-square test
Use it for categorical data.
It checks whether two variables are related or independent.
Example: checking whether gender and preference for a subject are related.

Z-test
Use it when you know the population standard deviation.
It is used for large samples.
Example: checking whether the average height of students in a school is different from the known average height of all students in the country.

AB-testing : A/B testing is a method of comparing two versions to find out which one gives better results.
