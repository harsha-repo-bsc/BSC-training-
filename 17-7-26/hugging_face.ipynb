```python
import torch
from transformers import AutoTokenizer, AutoModelForTokenClassification

# 1. Define a new sentence for NER
# This is the input text that we want to analyze for named entities.
text = "Barack Obama was born in Honolulu, Hawaii and later became the 44th President of the United States."

# 2. Load the tokenizer and model
# We load a pre-trained tokenizer and model specifically fine-tuned for Named Entity Recognition (NER).
# 'dslim/bert-base-NER' is a BERT-based model that has been trained on the CoNLL-2003 dataset
# for NER, which identifies persons (PER), organizations (ORG), locations (LOC), and miscellaneous (MISC).
tokenizer = AutoTokenizer.from_pretrained('dslim/bert-base-NER')
model = AutoModelForTokenClassification.from_pretrained('dslim/bert-base-NER')

# 3. Tokenize the new text
# The tokenizer converts the input text into a format that the model can understand.
# 'return_tensors='pt'' specifies that the output should be PyTorch tensors.
# The output 'inputs' will contain 'input_ids' (numerical representations of tokens)
# and 'attention_mask' (to differentiate real tokens from padding tokens).
inputs = tokenizer(text, return_tensors='pt')

# 4. Get model outputs
# We pass the tokenized inputs to the model to get its predictions.
# The 'outputs' object contains 'logits', which are the raw, unnormalized scores
# for each possible label for each token.
outputs = model(**inputs)

# 5. Extract predictions
# 'torch.argmax' is used to get the index of the highest logit score for each token along 'dim=2'.
# 'dim=2' refers to the dimension containing the scores for each label.
# This results in a tensor 'predictions' where each value is the predicted label ID for a token.
predictions = torch.argmax(
    outputs.logits,
    dim=2
)

# 6. Map predictions to labels
# We convert the numerical label IDs from 'predictions' back into human-readable label names
# using the model's configuration ('model.config.id2label').
# The 'predictions.view(-1)' flattens the tensor to make it easier to iterate through each token's prediction.
predicted_labels = []
predictions_flat = predictions.view(-1)

for prediction in predictions_flat:
  label_value = model.config.id2label[prediction.item()] # '.item()' extracts the scalar value from the tensor.
  predicted_labels.append(label_value)

# 7. Convert input IDs to human-readable tokens
# We convert the numerical input IDs back to their original word tokens using the tokenizer's 'convert_ids_to_tokens' method.
# '[0]' is used because 'input_ids' is typically batched, and we want the first (and only) sequence.
token_words = tokenizer.convert_ids_to_tokens(
    inputs['input_ids'][0]
)

# 8. Print token-label pairs
# This loop iterates through the original tokens and their corresponding predicted NER labels,
# displaying them in a clear, readable format.
# 'zip' is used to pair each token with its predicted label.
print(f"Text: {text}")
print("Named Entities:")
for word, predicted_label in zip(token_words, predicted_labels):
  print(f"{word} -> {predicted_label}")

# Expected Output Explanation:
# - [CLS] and [SEP]: Special tokens used by BERT, representing 'Classifier' (start of sentence) and 'Separator' (end of sentence). They are typically labeled 'O' (Outside).
# - B-PER: Beginning of a Person entity.
# - I-PER: Inside of a Person entity.
# - B-LOC: Beginning of a Location entity.
# - I-LOC: Inside of a Location entity.
# - O: Outside of any named entity.
# Notice how 'Barack' is 'B-PER' and 'Obama' is 'I-PER', indicating they form a single 'Person' entity.
```
