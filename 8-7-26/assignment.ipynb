1. What does API stand for, and what is its primary purpose in software communication?
2. Using the restaurant analogy from the session, explain the role of the “waiter” in an API request.

4. Why do companies prefer using existing APIs (for maps, payments, login, etc.) instead of building

API stands for Application Programming Interface; its primary purpose is to define rules/endpoints for apps to communicate and exchange data over the network.

In the restaurant analogy, the waiter takes your order (API request), sends it to the kitchen/server, and brings back the result (API response).

Google Maps uses map APIs (typically REST APIs) to fetch locations/routes; PayPal uses payment APIs (often REST APIs or webhooks/event APIs) for transactions.

Companies use existing APIs to save time and cost, rely on proven reliability/security, and benefit from standardized integrations instead of rebuilding complex features from scratch

5. Differentiate between the roles of a “Client” and a “Server” in client-server architecture, with one 
example of each.
6. What is the purpose of DNS, and why can't computers communicate using domain names alone?
7. Briefly explain how data physically travels across the internet in packets before reaching its 
destination.

Client sends requests (e.g., a browser/app asking for a page) while the server receives requests and sends back responses (e.g., web server returning HTML data). Example: Chrome is the client and a website’s server is the server.

DNS translates domain names (example.com) into IP addresses so computers know where to connect; domain names alone aren’t usable because routing happens using IPs, not names.

Data is split into packets; each packet travels through routers across the internet and may take different paths, then gets reassembled at the destination device to form the original message.

8. Define HTTP and explain why it is described as a “protocol.”
9. List the five components that make up an HTTP request.
10. What is the difference between response.status_code and response.json() when reading an API 
response?
11. Using the pizza-ordering analogy, match each part of an order (URL, Method, Headers, Body, 
Authentication) to its real-world equivalent.

HTTP is the Hypertext Transfer Protocol for transferring data between a client (browser/app) and a server. It’s called a “protocol” because it defines the exact rules/format for requests and responses.

HTTP request components: Method, URL (path), Headers, Body, and Authentication/credentials (often via Authorization header).

response.status_code tells the HTTP result (e.g., 200/404) without parsing content, while response.json() parses the response body into JSON data you can use.

URL = restaurant/address, Method = order type (pizza style), Headers = special instructions (size, allergies), Body = the pizza details you request, Authentication = your membership/ID to allow ordering.

12. Break down the URL https://api.openai.com/v1/chat/completions into its four components.
13. What is an API endpoint? Give an example of two different endpoints for a shopping application.
14. Match each HTTP method — GET, POST, PUT, PATCH, DELETE — to its correct purpose.
15. Explain the key difference between PUT and PATCH when updating a resource.

https://api.openai.com/v1/chat/completions
Components: scheme=https, domain=api.openai.com, path=/v1/chat/completions (endpoint resource=chat/completions, version=v1).

API endpoint is a specific URL/path where an API action happens. Example shopping app endpoints: /products (browse/search) and /cart/add (add item to cart).

GET=read, POST=create, PUT=replace/update fully, PATCH=partial update, DELETE=remove.

PUT replaces the entire resource with the new one, while PATCH updates only the fields you send (partial update).

16. What do the status codes 401 and 403 indicate, and how are they different from each other?
17. Why is a 429 status code returned, and what should a well-designed application do when it receives 
one?
18. What is the purpose of the “Authorization: Bearer <API_KEY>” header?
19. List three golden rules for handling API keys securely, as covered in the session.

401 Unauthorized means you’re not authenticated (missing/invalid login or token). 403 Forbidden means you’re authenticated but don’t have permission to access the resource.

429 Too Many Requests is returned when you exceed rate limits. A well-designed app should slow down, respect Retry-After, and retry/backoff instead of hammering the API.

“Authorization: Bearer <API_KEY>” sends the API key/token so the server can authenticate and authorize the request.

Golden rules: never hardcode keys in code/notebooks, store them in environment variables/secret managers, and don’t share/expose keys in logs or client-side code.

20. Why is JSON considered the “standard format” for API responses?
21. Explain one key difference between a Python dictionary and JSON.
22. Convert the following Python dictionary into valid JSON format: {"name": "Employee", 
"department": "IT", "active": True}

JSON is considered standard because it’s lightweight, text-based, easy to read/write, and widely supported by languages and APIs for consistent data exchange.

Python dict uses Python-native types/booleans (True/False) and allows non-JSON types; JSON requires specific JSON types (true/false, double quotes) and a strict format.

26. What is the advantage of using a requests.Session() object over calling requests.get() repeatedly?

Read data: requests.get(); Create new data: requests.post(); Delete data: requests.delete().

timeout limits how long the request can hang; it’s best practice to prevent blocking forever and to fail fast on network/API issues.

try:
response = requests.get(URL, timeout=10)
response.raise_for_status()

except requests.exceptions.RequestException as e:
print('request failed:', e)

requests.Session() reuses connections and shared headers/cookies, improving performance and consistency vs repeated requests.get() calls.

30. In your own words, summarize the complete journey of an API call — from client request to server 
response — using the concepts covered in this session.

You type a prompt in the client → it’s sent over HTTPS to the service endpoint → the request hits the API gateway and then the AI model → the model generates tokens → the service streams/returns the final response back to your client.

Use least-privilege keys, store keys in env/secret manager (not code), use HTTPS + validate responses, follow rate limits/retries (timeouts/backoff) and log safely without exposing secrets.

Risk: the leaked key can be used for unauthorized calls/charges. Correct approach: rotate/revoke the key immediately and store it in environment variables/secret manager, never in GitHub.

Client builds request (URL + method + headers/auth + optional body) → sends it as an HTTP request → server routes/handles it → returns an HTTP response with status_code and JSON data (response body) → client parses/uses the data.
