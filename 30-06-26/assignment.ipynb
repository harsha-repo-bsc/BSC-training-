• Less than 8 characters → Weak
• 8–12 characters → Medium
• Above 12 characters → Strong

```python

try :
    password = input('Enter your password :')
    if password == '':
        raise Exception('Password cannot be empty')
except Exception as e:
    print(e)
else:
    pass_len = len(password)

    if pass_len < 8:
        print('Weak Password')
    elif pass_len >= 8 and pass_len < 12:
        print('Moderate Password')
    elif pass_len >= 12:
        print('Strong Password')

finally :
    print('Password validation completed')
    print('Your password is :', password)
```

Q16. Encapsulation — Bank Account
Create a class BankAccount where the balance is kept private. Create methods deposit(), withdraw(),
and check_balance().

```python
class Bank_account:
    def __init__(self, initial_balance = 0):
        self.__balance = initial_balance

    def balance_check(self):
        return self.__balance

    def credit(self,credit):
        self.__balance +=credit
        return self.__balance

    def debit(slef,debit):
        if self.__balance <= debit:
            return slef.__balance-debit
        else:
            return 'insufficient funds'

user = Bank_account()
print(user.credit(100))
print(user.balance_check())
print(user.credit(200))
user.balance_check()
```

Q19. Abstraction — Model
Create an abstract class Model with an abstract method train(). Create two child classes that
implement it

```python
from abc import ABC, abstractmethod

class Properties(ABC):
  
    @abstractmethod
    def engine(self):
        """Talks about the engine"""


    @abstractmethod
    def design(self):
        """Talks about the design"""

class CarProperties(Properties):
    @property
    def engine(self):
        return "Here I am defining the engine properties"

    @property
    def design(self):
        return "Here I am discussing design methods"

c = CarProperties()
print(c.engine)
print(c.design)
```
