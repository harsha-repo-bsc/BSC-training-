```python
class car :
    
    def __init__(self , brand , color ):       # magic words or dunder methods ()
        self.brand  = brand 
        self.color = color 

    def display_car(self):
        print('CAR Brand : ' , self.brand)
        print('CAR Color : ' , self.color)
        print( '**'*20)

bmw  = car('BMW' , 'Black')
bmw.display_car()

# object creation and method calling 

maruthi = car('Maruthi' , 'White')
maruthi.display_car()


```

```python
# abstraction : hiding of implementation details(internal) and showing only functionality(name ) to the user is called abstraction.

from abc import abstractmethod

class car() :

    @abstractmethod
    def new_feature(self):
        print('New Feature : Auto Parking')



# give me a perfect samll example for abstraction in python 

class car() :
    
    @abstractmethod
    def new_feature(self):
        pass

# car class is abstract class and new_feature is abstract method.
# lets define a child class of car class and implement the abstract method new_feature.

class BMW(car) :
    # use @ property decorator to implement the abstract method new_feature.

    @property   # its a decorator which is used to define a method as a property of the class.
    def new_feature(self):
        print('New Feature : Auto Parking')


# so while creating an object of BMW class we can access the new_feature method as a property of the class.

bmw = BMW()
bmw.new_feature   # this will call the new_feature method of BMW class and print the


```

```python
# inheritance : when a class is derived from another class then the derived class is called child class and the base class is called parent class. The child class inherits the properties and methods of the parent class.

class car :
    
    def engine(self):
        print('Engine : 4 Cylinder')
    
    def wheels(self):
        print('Wheels : 4')

class BMW(car) :
    pass

bmw = BMW()
bmw.engine()  # this will call the engine method of car class and print the engine details.
bmw.wheels()  # this will call the wheels method of car class and print the wheels details. 

# the BMW class is inheriting the properties and methods of car class.
```

```python
# polymorphism : the ability of an object to take many forms is called polymorphism. In python, polymorphism is achieved by method overriding and operator overloading.


class ai_model :

    def predict(self):
        print('Predicting the output')

class ai_model_v2(ai_model) :
    def predict(self):
        print('Predicting the output using ai_model_v2')

# how polymorphism is achieved in python using method overriding.


```

```python
# magic methods  : are the special methods which are used to perform operator overloading and method overriding. 
# The magic methods are defined by double underscores before and after the method name. 
# For example, __init__ is a magic method which is used to initialize the object of a class.


# __init__
# __str__
# __add__
# __len__
# __repr__
# __eq__


class student:

    def __init__(self, name , age):
        self.name = name
        self.age = age 

    def display(self):
        print('Name : ' , self.name)
        print('Age : ' , self.age)


student1 = student('John' , 20)
student1.name       # how come the name variable is accessible outside the class. This is because the name variable is a public variable and can be accessed outside the class.
```

```python
# __str__ : is a magic method which is used to return a string representation of the object of a class.
#  This method is called when we print the object of a class.

class student :

    def __init__(self,name , age):
        self.name = name
        self.age = age

    def __str__(self):
        return 'Name : ' + self.name 

student1 = student('John' , 20)
print(student1)   # this will call the __str__ method of student class and print
```

```python
# ai sentiment analysis for online shopping 


class sentiment_analyer():

    positive = ['good' , 'excellent' , 'awesome' , 'fantastic' , 'great']
    negative = ['bad' , 'poor' , 'terrible' , 'awful' , 'worst']


    def __init__(self, review):
        self.review = review 
    
    def sentiment_analysis(self):
        try : 
            self.review = self.review.lower()
            if self.review == "":
                raise Exception('Review cannot be empty')
            if self.review.isdigit():
                raise Exception('Review cannot be a number')

        except ValueError:
            print('Error : Invalid input')
        except Exception as e:
            print('Error : ' , e)
        positive_count = 0
        negative_count = 0

        for word in self.review.split():
            if word in self.positive:
                positive_count += 1
            elif word in self.negative:
                negative_count += 1
        if positive_count > negative_count:
            print('Sentiment : Positive')
        elif negative_count > positive_count:
            print('Sentiment : Negative')
        else:
            print('Sentiment : Neutral')
        
    def display(self):
        print('YOUR Review : ' , self.review)

review = input('Enter your review : ')

a1 = sentiment_analyer(review)
a1.display()
a1.sentiment_analysis()

```
