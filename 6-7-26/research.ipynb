An Application Programming Interface (API) is a set of rules and protocols that enables different software applications to communicate and exchange data.

Why APIs Are Used

Access live data
Automate data collection
Integrate multiple applications
Reduce manual work
Enable real-time analytics

What is an Endpoint?
An endpoint is a specific URL where an API can be accessed.
Example
Suppose we use the JSONPlaceholder API:
https://jsonplaceholder.typicode.com/posts


HTTP Methods
When working with APIs, different methods are used to communicate with the server.
GET -requests.get(url)
Used to retrieve data.
POST - requests.post(url, json=data)
Used to create new records.

PUT - requests.put(url, json=data)
Used to update existing data.

DELETE - requests.delete(url)
Used to remove data.


What Are Headers?
Headers provide additional information about the request.
They tell the server:

Who is making the request
What format is expected
Authentication details

headers = {
    "Accept": "application/json",
    "User-Agent": "Python Script"
}

response = requests.get(url, headers=headers)

| Header        | Purpose                  |
| ------------- | ------------------------ |
| Accept        | Expected response format |
| Content-Type  | Data format sent         |
| Authorization | Authentication token     |
| User-Agent    | Client information       |



Authentication
Many APIs protect their data and require authentication.
Authentication verifies the identity of the user requesting the data.

Basic Authentication - Username + Password
- requests.get(url, auth=("username", "password"))

API Key Authentication
User is assigned a unique key.
Example:api_key = "YOUR_API_KEY"

url = f"https://api.service.com/data?apikey={api_key}"

Token Authentication
A secure access token is provided.
Example: headers = {
    "Authorization": "Bearer YOUR_TOKEN"
}


Example: Fetching Data from a Public API


    import requests

    url = "https://jsonplaceholder.typicode.com/users"

    response = requests.get(url)

    print(response.status_code)
    print(response.json())
