```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import logging

for handler in logging.root.handlers[:]:
  logging.root.removeHandler(handler)

logging.basicConfig (              # Saving Severity Level
    level = logging.INFO,
    format = '%(asctime)s | %(levelname)s | %(message)s',
    datefmt = '%Y-%m-%d %H:%M:%S',
    force = True
)
logging.info("Application Started Successfully")

data = {
    "Experience": [1, 2, 3, 4, 5, 6],
    "Salary": [25000, 45000, 60000, 75000, 90000, 105000]
}


df = pd.DataFrame(data)

X = df[["Experience"]]
y = df["Salary"]

logging.info(f"Dataset Shape: {df.shape}")
```

```python
model = LinearRegression()
model.fit(X,y)

logging.info(f'Intercept: {model.intercept_}')
logging.info(f'Coefficient: {model.coef_[0]}')

prediction = model.predict([[7]])

logging.info(f'Prediction completed successfully')
print('predicted_salary',prediction[0])

logging.info("Application finished  Successfully")
```

Logging is the process of recording events, activities, and system information while an application is running. By incorporating location information and GPS coordinates into logs, developers can track where specific events occur, making applications more useful for monitoring, analysis, and troubleshooting.

 Location Information and GPS Coordinates in Python Logging
Python's logging module can be enhanced to include location details such as:

Latitude and Longitude
City or Region
GPS Accuracy
Device Location Status

These details can be collected from:

GPS-enabled devices
Mobile applications
IoT sensors
External location APIs

2. Role of logging.basicConfig()
logging.basicConfig() is used to configure the logging system.
It allows developers to define:

Log file name
Log level (INFO, WARNING, ERROR)
Log message format
Timestamp format

Benefits:

Standardizes log messages.
Stores logs in files.
Adds timestamps automatically.
Helps in debugging and monitoring.

3. Capturing Contextual Information
Logging can capture various contextual details along with application events.

4. Real-World Applications of Location-Aware Logging
Fleet Management
Transportation companies track vehicle movements using GPS-enabled logs.
Benefits:

Route monitoring
Driver tracking
Fuel optimization

Delivery Tracking
Companies like courier and food delivery services use location logs to:

Track delivery personnel
Record delivery locations
Estimate arrival times

IoT Systems
IoT devices often include GPS sensors and logging capabilities.
Examples:

Smart agriculture
Asset tracking
Environmental monitoring

Logs help identify where devices are located and what activities they perform.
