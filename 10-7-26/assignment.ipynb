```python
# Q47
def relu(values):
    return [max(0, v) for v in values]

values = [-3, -1, 0, 4, 7]
print(relu(values))
```

Q50. For the Softmax calculation with raw scores [2.0, 1.0, 0.1]:

Step 1: Compute exponentials  
e^2 = 7.389  
e^1 = 2.718  
e^0.1 = 1.105  

Step 2: Add them  
Sum = 7.389 + 2.718 + 1.105 = 11.212  

Step 3: Probability of Class 1  
P(Class 1) = e^1 / Sum = 2.718 / 11.212 ≈ 0.242

So, the probability of Class 1 is approximately 0.242 or 24.2%.
